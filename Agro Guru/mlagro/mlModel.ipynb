```python
import numpy as np
import pandas as pd
```

```python
original_df = pd.read_csv("Crop_recommendation.csv")
crop = pd.DataFrame(original_df)
crop
```

```python
x=crop.drop('label',axis=1)
x
```

```python
y=crop['label']
y
```

```python
crop.dropna(inplace=True)
crop
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
crop['label']=lb.fit_transform(crop['label'])
crop
```

```python
x=crop.drop('label',axis=1)
x
y=crop['label']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

```

```python
from sklearn.ensemble import RandomForestClassifier
classifier= RandomForestClassifier(n_estimators= 100,max_depth=5, criterion="entropy")
classifier.fit(x_train, y_train)
x_pred=classifier.predict(x_train)
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
mean=mean_squared_error(y_train,x_pred)
r2=r2_score(y_train,x_pred)*100
r2
```

```python
mean=mean_squared_error(y_test,y_pred)*100
r2=r2_score(y_test,y_pred)
pred=classifier.predict(x_test)
pred
```

```python
ac=accuracy_score(y_test,y_pred)
ac*100
```
